### 시군구 단위 자료 1차 가공
1. 2022 선거 데이터 처리

```python
voting = pd.DataFrame(pd.read_excel("voting_data_2022.xlsx"))
```

```python
voting.iloc[0:,4:] = voting.iloc[0:,4:].apply(lambda x : pd.to_numeric(x.str.replace(r',','')))
new_vote_2022 = voting[voting['읍면동명']=="합계"]
new_vote_2022['outtotal_2022'] = new_vote_2022.iloc[:,7:18].sum(axis=1)
```

```python
new_vote_2022.iloc[:,18] = new_vote_2022.iloc[:,18].replace(0, np.nan)
new_vote_2022['outratio_2022'] = new_vote_2022.iloc[:,21]/new_vote_2022.iloc[:,18]
new_vote_2022.head()
```

2. 2018 선거 데이터 처리

```python
voting2 = pd.DataFrame(pd.read_excel("voting_data_2018.xlsx"))
```

```python
new_vote_2018 = voting2[voting2['읍면동명'] == '계']
new_vote_2018['outtotal_2018'] = new_vote_2018.iloc[:,9:16].sum(axis=1)
new_vote_2018.iloc[:,17] = new_vote_2018.iloc[:,17].replace(0, np.nan)
new_vote_2018['outratio_2018'] = new_vote_2018.iloc[:,20]/new_vote_2018.iloc[:,17]
```

3. 데이터 병합

```python
new_vote_2022.rename(columns={'구시군':'구시군명'},inplace=True)
```

```python
new_vote_2022['구시군명']
```

```python
new_vote_2018['구시군명']
```

```python
new_vote_2022.rename(columns={'시도':'시도명'},inplace=True)
```

```python
vote_2018_2022_t = pd.merge(new_vote_2018, new_vote_2022, on = ["구시군명", "시도명"])
```

```python
vote_2018_2022_t.head()
```

```python
vote_2018_2022_t.columns
```

```python
vote_2018_2022_t.rename(columns={'투표수_x':'2018_total_vote', '투표수_y' : '2022_total_vote', '선거인수_x' : '2018_voter_all', '선거인수_y' : '2022_voter_all'},inplace=True)
```

```python
vote_2018_2022_sigungu = vote_2018_2022_t[["시도명","구시군명",'2018_total_vote',"outtotal_2018", "outratio_2018",'2022_total_vote', "outtotal_2022","outratio_2022",'2018_voter_all','2022_voter_all']]
```

```python
vote_2018_2022_sigungu['2018_vote_ratio'] = vote_2018_2022_sigungu['2018_total_vote']/vote_2018_2022_sigungu['2018_voter_all']
vote_2018_2022_sigungu['2022_vote_ratio'] = vote_2018_2022_sigungu['2022_total_vote']/vote_2018_2022_sigungu['2022_voter_all']
```

```python
vote_2018_2022_sigungu.head()
```

최종적으로 저장

```python
vote_2018_2022_sigungu.to_excel("sigungu_vote_data.xlsx", index = False)
```

3. 교통 데이터 전처리

```python
test_data_for_tran = pd.read_excel("접근가능인구비율_시군구.xlsx")[["HDCD_SD_NM", "HDCD_SGG_NM", "Mode_NM","Faci_NM","30분 이내"]]
test_data_for_tran = test_data_for_tran[test_data_for_tran["Mode_NM"]=="대중교통/도보"][["HDCD_SD_NM", "HDCD_SGG_NM","Faci_NM","30분 이내"]]
```

```python
test_data_for_tran.columns
```

```python
test_data_for_tran["index"] = test_data_for_tran["HDCD_SD_NM"].astype('str') + "_" + test_data_for_tran["HDCD_SGG_NM"].astype('str')
```

```python
test_data_for_tran.head()
```

```python
test_data_for_tran = test_data_for_tran.drop_duplicates(subset=['index', 'Faci_NM'])
test_data_for_tran = test_data_for_tran.pivot(index = 'index', columns="Faci_NM", values="30분 이내")
```

```python
test_data_for_tran.head()
```

```python
test_data_for_tran.to_excel("data_for_tran.xlsx")
```
