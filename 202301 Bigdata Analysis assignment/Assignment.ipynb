3. Assignment 6

```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
import statsmodels.formula.api as smf
```

```python
ass6  = pd.read_csv("./ID-8.txt", sep="\t", header=None)
ass6.dropna(inplace=True)
ass6.columns = ["location", "household_size" ,"age", "sex" , "maristat" ,"school", "school2", "worktype", "wage" ,"jobtype", "englishtime", "english", "union","certi","yrsschl"]
```

```python
ass6.head()
```

```python
ass6['sex'].replace(2,0,inplace=True)
```

```python
ass6['logwage'] =  np.log(ass6['wage'])
```

```python
np.log(ass6['wage'])
```

```python
ass6['logwage']
```

```python
ass6['sex']
```

```python
np.isinf(ass6['logwage']).sum()
```

```python
ass6.replace([np.inf, -np.inf], np.nan, inplace=True)
```

```python
ass6.dropna(inplace=True)
```

```python
ass6.head()
```

```python
ass6.describe()
```

```python
anle_6 = smf.ols(formula="logwage ~ sex",data = ass6).fit()
print(anle_6.summary())
```

```python
anle_6_2 = smf.ols(formula="logwage ~ sex",data = ass6).fit(cov_type='HC2')
print(anle_6_2.summary())
```

```python
anle_6_3 = smf.ols(formula="logwage ~ sex + age + yrsschl",data = ass6).fit()
print(anle_6_3.summary())
```

```python
anle_6_4 = smf.ols(formula="logwage ~ sex*yrsschl",data = ass6).fit()
print(anle_6_4.summary())
```

```python
ass6['yrsschl'].mean()
```

```python
anle_6_5 = smf.gls(formula="logwage ~ sex + age + yrsschl + 0",data = ass6).fit()
print(anle_6_5.summary())
```

```python
for i in ['logwage','sex', 'age','yrsschl'] :
    ass6[i] = ass6[i]/np.sqrt(ass6['age'])
```

```python
ass6['reverseage'] = 1/np.sqrt(ass6['age'])
```

```python
anle_6_6 = smf.ols(formula="logwage ~ sex + age + yrsschl + reverseage + 0",data = ass6).fit()
print(anle_6_6.summary())
```

Assignment 7

```python
ass7 = pd.read_csv('./growth-inf.txt', sep="\t")
```

```python
ass7.columns = ["growth_rate", "inflation_rate"]
```

```python
ass7.head()
```

```python
ass7_reg = smf.ols(formula='growth_rate ~ inflation_rate', data = ass7).fit()
print(ass7_reg.summary())
```

```python
residual_t = ass7_reg.resid
```

```python
len(residual_t)
```

```python
residual_t.head()
```

```python
residual_t_1 = residual_t.shift()
```

```python
residual_t_1.head()
```

```python
residuals = pd.DataFrame({"residualt" : residual_t, "residualt1" : residual_t_1}).dropna()
```

```python
residuals.head()
```

```python
ar_1 = smf.ols('residualt ~ 0 + residualt1', data = residuals).fit()
print(ar_1.summary())
```
